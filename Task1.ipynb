

# **Iris Dataset – Data Loading, Inspection, and Visualization**

## **1. Data Loading**

* The **Iris dataset** was loaded using `seaborn.load_dataset()`.
* A pandas DataFrame named **df** was created.

---

## **2. Basic Data Inspection**

To understand the dataset structure, the following were examined:

* `df.shape` – dimensions of the dataset
* `df.columns` – list of columns
* `df.head()` – first few rows
* `df.info()` – data types and non-null counts
* `df.describe()` – summary statistics for numerical features

---

## **3. Data Visualization**

Several visualizations were generated using **seaborn**:

### **Scatter Plot**

* Scatterplot of **petal length vs petal width**, colored by **species**,
  to visually observe class separation among iris species.

### **Pair Plot**

* A **pair plot** showing:

  * Relationships between all pairs of numerical features
  * Individual feature distributions
  * Coloring by **species** for better comparison

### **Histograms and Box Plots**

* For each numerical feature:

  * **Sepal length**
  * **Sepal width**
  * **Petal length**
  * **Petal width**
* Histograms were used to check distributions.
* Box plots were used to identify outliers and variation within each feature.

---
